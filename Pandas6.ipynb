```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna() #drops rows 3,4,5 having NaN values
```

```python
df.dropna(axis = 'index', how = 'any')
#both gives same result ...here how = 'any' means if NaN value at any column in a row then drop that row
```

```python
df.dropna(axis = 'index', how = 'all')
#here how = 'all' means drop row if all column having NaN value
```

```python
df.dropna(axis = 'columns', how = 'all')
#drop column if all values are NaN
```

```python
df.dropna(axis = 'columns', how = 'any')
#all columns deleted as each column having at least one NaN value
```

```python
df.dropna(axis = 'index', how = 'any', subset = ['email'])
# if there is email then it will not drop that row
```

```python
df.dropna(axis = 'index', how = 'any', subset = ['email','last'])
#if row contain both email and last then it will not drop
```

```python
df.replace('Missing',np.nan)
```

```python
#How to handle dynamic NN values
df.replace(['Missing','NA'],np.nan ,inplace = True)
df
```

```python
df.isna()
#NaN vales 
```

```python
df.fillna("MISSING")
```

```python
df.fillna(0)
```

```python
df.dtypes
#here all values are objects/string   so we can not apply aggregate function on that
```

```python
#convert string type in to float/int
df.age = df.age.astype(float)
```
